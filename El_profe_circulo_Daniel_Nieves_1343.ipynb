## formula para calcular Area

## formula para calcular perimetro

```python
# calcular area y perimetro de un circulo
# variables A, P, r,
# formulas A = pi*r^2 y P = 2 * pi * r
#Capturar el valor de r
import math
print("Calcular area y perimetro de un circulo")
r = float(input("ingrese el valor de r: "))
A = math.pi*r**2
P = 2*math.pi*r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Daniel Nieves NL = 41")
```
